```python
import numpy as np
import stochoptim.scengen.tree_structure as ts
```

# Build a Tree Structure

### Tree Structure `from_bushiness`

```python
tree = ts.from_bushiness([3,2,2])
tree.plot()
```

### Tree Structure `from_topology`

```python
tree = ts.from_topology([[1,2,4], [2,3]])
tree.plot()
```

### Tree Structure `from_recurrence`

```python
tree = ts.from_recurrence(max_level=5, init=3, recurrence={1: [2], 2: [1,2], 3: [1,2,3]})
tree.plot()
```
